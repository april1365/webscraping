```python
firstjob = BS(response.text)
```

```python
firstjob_description = firstjob.find("div", class_="content").find("p").get_text(strip=True)
print(firstjob_description)
```

```python
def get_description(url):
    response = requests.get(url)
    soup = BS(response.text, 'html.parser')
    description = soup.find("div", class_="content").find("p").get_text(strip=True)
    return description
```

```python
description = get_description('https://realpython.github.io/fake-jobs/jobs/television-floor-manager-8.html')
print(description)
```

```python
job_postings['description'] = job_postings['url'].apply(get_description)
job_postings
```
